```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns

df = pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data), dtype=object)
```

```python
df
```

```python
df['a'][1]
```

```python
df = pd.read_csv(StringIO(data), dtype={
    'b': int,
    'c': float,     # or np.float64
    'a': 'Int64'
})
```

```python
print(df)
```

```python
print(df['a'][1])
```

```python
## check the datatype

print(df.dtypes)
```

```python
## Index columns and training delimiters

data = ('index,a,b,c\n'
        '4,apple,bat,5.7\n'
        '8,orange,cow,10\n')
```

```python
print(pd.read_csv(StringIO(data)))
```

```python
print(pd.read_csv(StringIO(data), index_col=0))
```

```python
## Combining usecols and index_col

data = ('index,a,b,c\n'
        '4,apple,bat,5.7\n'
        '8,orange,cow,10\n')
```

```python
print(pd.read_csv(StringIO(data), usecols=['b', 'c'], index_col=False))
```

```python
## Quoting and Escape Characters. Very useful in NLP

data = 'a,b\n "hello, \\"Bob\\", nice to see you",5'
```

```python
print(pd.read_csv(StringIO(data), escapechar='\\'))
```

MatplotLib Tutorial

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Sample Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib
## plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
#plt.savefig('Test.png')

plt.show()
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*', linestyle='dashed', markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Diagram')
plt.show()
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'go')

plt.show()
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y)
plt.show()
```

```python
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave")

# Plot the points using set_matplotlib_close
plt.plot(x,y,'g--')
plt.show()
```

```python
# Subplot()

x=np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos,'.')
plt.title('Cosine')

# Show the figure.
plt.show()
```

```python
## Bar Plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0) #explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
